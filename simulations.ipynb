# Simulations 

This document is meant to show some simple simulations with the monopoly.py file. 

```python
import monopoly
import pandas as pd
import matplotlib as plt 
%matplotlib inline
```

```python
res = [] 
for i in range(100): 
    board = monopoly.Board()
    board.start_game(100)
    res.append([p.cash_left for p in board.players])
```

```python
cols = ['player', 'buyall', 'buyfrom', 'buybetween']
df = pd.DataFrame(res, columns = cols)
_ = df.hist(figsize=(10,8), bins = 30)
```

It seems that `buyall` might not be the best tactic. Next it may be interesting to have a look at the time series during a single game.
